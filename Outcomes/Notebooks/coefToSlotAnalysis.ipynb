```python
# Imports
import os
import glob
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read CSVs
innerProdCoefAsSlotDF = pd.read_csv('../timeCSVs/innerProdCoefToSlotTimes.csv').drop([' ringDim', ' total_time',' value '], axis=1)

sumCoefAsSlotDF = pd.read_csv('../timeCSVs/sumCoefToSlotTimes.csv').drop([' ringDim', ' total_time',' value '], axis=1)

sumCoefDF = pd.read_csv('../timeCSVs/sumCoefTimes.csv').drop([' ringDim', ' total_time',' value '], axis=1)

innerProdCoefDF = pd.read_csv('../timeCSVs/innerProdCoefTimes.csv').drop([' ringDim', ' total_time',' value '], axis=1)

sumSlotDF = pd.read_csv('../timeCSVs/sumSlotTimes.csv').drop([' ringDim', ' total_time',' value '], axis=1)

innerProdSlotDF = pd.read_csv('../timeCSVs/innerProdSlotTimes.csv').drop([' ringDim', ' total_time',' value '], axis=1)
```

```python
sumCoefAsSlotDF = sumCoefAsSlotDF.groupby(['type'], as_index=False).mean()
innerProdCoefAsSlotDF = innerProdCoefAsSlotDF.groupby(['type'], as_index=False).mean()

sumCoefDF = sumCoefDF.groupby(['type'], as_index=False).mean()

innerProdCoefDF = innerProdCoefDF.groupby(['type'], as_index=False).mean()

sumSlotDF = sumSlotDF.groupby(['type'], as_index=False).mean()

innerProdSlotDF = innerProdSlotDF.groupby(['type'], as_index=False).mean()
```
