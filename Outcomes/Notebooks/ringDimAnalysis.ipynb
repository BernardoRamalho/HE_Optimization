```python
# Imports
import os
import glob
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import Data From CSV
small_ring_df = pd.read_csv("../timeCSVs/smallRingDim.csv")
small_ring_df.drop([' decryption', ' plaintext', ' value'], axis=1,inplace=True)

big_ring_df = pd.read_csv("../timeCSVs/bigRingDimTimes.csv")
big_ring_df.drop([' decryption', ' plaintext', ' value'], axis=1,inplace=True)

big_elements_df = pd.read_csv("../timeCSVs/ringDimOptimization.csv")
big_elements_df.drop([' decryption', ' plaintext', ' value'], axis=1,inplace=True)


ring_df = pd.concat([small_ring_df, big_ring_df, big_elements_df], ignore_index=True) 

ring_df.head()
```
